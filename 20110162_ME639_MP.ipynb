```python
from turtle import width
import numpy as np
import math
import matplotlib.pyplot as plt
import cv2
import scipy
from render import Renderer
```

Defining a class object

```python
class object(Renderer):
    def __init__(self, recordLocation=None):
        super().__init__(recordLocation=recordLocation)
        
        self.its = 0
        
        self.x0 = 50    # X-reference coordinate for the base joint
        self.y0 = 500   # Y-reference coordinate for the base joint
        self.l1 = 225   # The length of the first arm
        self.l2 = 225   # The length of the second arm
        
        self.m1 = 5     # The mass of the first arm
        self.m2 = 10    # The mass of the first arm
        
        self.points = set()     # Set for storing the points in a set
    
    # getInfo function for rounding off the values of the calculated data points
    def getInfo(self):
        info = {
            'x1' : round(self.x1, 4),
            'y1' : round(self.y1, 4),
            'x2' : round(self.x2, 4),
            'y2' : round(self.y2, 4)
        }
        return info
    
    # plot functionto plot the values on a plot to give the random geometry
    def plot(self):
        data = np.array(list(self.points))
        plt.scatter(data[:,4],data[:,5],)
        plt.show()
    
    # step function iterates to calculate the different data points 
    def step(self, x, y):
        
        # Taking the coordinates of the circle provided from the user
        self.x = x 
        self.y = y

        # Calculating the theta angle
        self.theta = math.acos((((self.x-self.x0)**2 + (self.y-self.y0)**2 - self.l1**2 - self.l2**2)/(2*self.l1*self.l2)))

        # Calculating q1 and q2 angle
        self.q1 = math.atan((self.y-self.y0)/(self.x-self.x0))-math.atan(self.l2*math.sin(self.theta)/(self.l1+self.l2*math.cos(self.theta)))
        self.q2 = self.q1 + self.theta
        
        # Calculating the end points of the first arm from the point(x0,y0)
        self.x1 = self.x0 + self.l1*math.cos(self.q1)
        self.y1 = self.y0 + self.l1*math.sin(self.q1)
        
        # Calculating the end points of the first arm from the point(x1,y1)
        self.x2 = self.x1 + self.l2*math.cos(self.q2)
        self.y2 = self.y1 + self.l2*math.sin(self.q2)
        
        self.its += 1
        
        # Adding the points in the coordinated in a set
        self.points.add((self.x1, self.y1, self.x2, self.y2, self.x, self.y))

    # draw function draws the image as per the user requirement on the basis of given coordinates
    def draw(self,image):
        for x1,y1,x2,y2,x,y in self.points:
            cv2.line(image,(self.x0,self.y0),(int(self.x1),int(self.y1)),(0,0,255),1)
            cv2.line(image,(int(self.x1),int(self.y1)),(int(self.x2),int(self.y2)),(0,255,0),1)
            cv2.circle(image,(int(x),int(y)),1,(255,0,0),-1)
        return image
```

```python
for i in range(628):
    xc.append(xc0 + r*math.cos(i*0.01))
    yc.append(yc0 + r*math.sin(i*0.01))
```

Creating an object to call the class object

```python
anim = object(recordLocation='anim.mp4')
```

for loop which runs and make the animation by continuously rendering the data points

Defining a class object

```python
class object(Renderer):
    def __init__(self, recordLocation=None):
        super().__init__(recordLocation=recordLocation)
        self.its = 0
        self.x0 = 150    # X-reference coordinate for the base joint
        self.y0 = 500   # Y-reference coordinate for the base joint
        self.l1 = 250   # The length of the first arm
        self.l2 = 250   # The length of the second arm
        self.m1 = 5     # The mass of the first arm
        self.m2 = 10    # The mass of the first arm
        self.points = set()     # Set for storing the points in a set

    # getInfo function for rounding off the values of the calculated data points
    def getInfo(self):
        info = {
            'x1' : round(self.x1, 4),
            'y1' : round(self.y1, 4),
            'x2' : round(self.x2, 4),
            'y2' : round(self.y2, 4)
        }
        return info
    
    # plot functionto plot the values on a plot to give the random geometry
    def plot(self):
        data=np.array(list(self.points))
        plt.scatter(data[:,4],data[:,5],)
        plt.show()
    
    # step function iterates to calculate the different data points
    def step(self, dt, xw, yw, xr, yr, fx, fy):
        g = 9.81 # Defining gravity

        # Taking the coordinates of the circle provided from the user
        self.x = xr + self.its*dt
        if xw!=self.x:
            self.y = ((yw - yr)/(xw - xr))*(self.x - xr)

        # Taking the forces along x and y directions as per the user input
        self.Fx = fx
        self.Fy = fy

        # Calculating the theta angle
        self.theta = math.acos((((self.x-self.x0)**2 + (self.y-self.y0)**2 - self.l1**2 - self.l2**2)/(2*self.l1*self.l2)))

        # Calculating q1 and q2 angle
        self.q1 = math.atan((self.y-self.y0)/(self.x-self.x0))-math.atan(self.l2*math.sin(self.theta)/(self.l1+self.l2*math.cos(self.theta)))
        self.q2 = self.q1 + self.theta
        
        # Calculating the end points of the first arm from the point(x0,y0)
        self.x1 = self.x0 + self.l1*math.cos(self.q1)
        self.y1 = self.y0 + self.l1*math.sin(self.q1)
        
        # Calculating the end points of the first arm from the point(x1,y1)
        self.x2 = self.x1+self.l2*math.cos(self.q2)
        self.y2 = self.y1+self.l2*math.sin(self.q2)

        # Calculating the torques for both the arms
        self.t1 = -self.Fx*self.l1*math.sin(self.q1) + self.Fy*self.l1*math.cos(self.q1) + self.m1*g*self.l1*math.cos(self.q1)/2 + self.m2*g*self.l1*math.cos(self.q1)
        self.t2 = -self.Fx*self.l2*math.sin(self.q2) + self.Fy*self.l2*math.cos(self.q2) + self.m2*g*self.l2*math.cos(self.q2)/2
        
        self.its += 1

        # Adding the points in the coordinated in a set
        self.points.add((self.x1, self.y1, self.x2, self.y2, self.x, self.y))
        print(self.t1)
        print(self.t2)
    
    # draw function draws the image as per the user requirement on the basis of given coordinates
    def draw(self,image):
        for x1,y1,x2,y2,x,y in self.points:
            cv2.line(image,(self.x0,self.y0),(int(self.x1),int(self.y1)),(0,0,255),1)
            cv2.line(image,(int(self.x1),int(self.y1)),(int(self.x2),int(self.y2)),(0,255,0),1)
            cv2.circle(image,(int(x),int(y)),1,(255,0,0),-1)
        return image
```

Defining a class object

```python
class object(Renderer):
    def __init__(self, recordLocation=None):
        super().__init__(recordLocation=recordLocation)
        self.its = 0
        self.x0 = 50    # X-reference coordinate for the base joint
        self.y0 = 500   # Y-reference coordinate for the base joint
        self.l1 = 225   # The length of the first arm
        self.l2 = 226   # The length of the second arm
        self.m1 = 5     # The mass of the first arm
        self.m2 = 10    # The mass of the first arm
        self.points = set()     # Set for storing the points in a set

    # getInfo function for rounding off the values of the calculated data points
    def getInfo(self):
        info = {
            'x1' : round(self.x1, 4),
            'y1' : round(self.y1, 4),
            'x2' : round(self.x2, 4),
            'y2' : round(self.y2, 4)
        }
        return info
    
    # plot functionto plot the values on a plot to give the random geometry
    def plot(self):
        data=np.array(list(self.points))
        plt.scatter(data[:,4],data[:,5],)
        plt.show()
    
    # step function iterates to calculate the different data points
    def step(self, xc, yc):
        g = 9.81    # Defining gravity
        k = 10      # Defining a constant which acts similar as a spring constant

        # Taking the coordinates of the circle provided from the user
        self.x = xc
        self.y = yc

        # Calculating the force required to spring back to the position we want the robotic arm to be
        self.Fx = k*(self.x - xc)
        self.Fy = k*(self.y - yc)

        # Calculating the theta angle
        self.theta = math.acos((((self.x-self.x0)**2 + (self.y-self.y0)**2 - self.l1**2 - self.l2**2)/(2*self.l1*self.l2)))

        # Calculating q1 and q2 angle
        self.q1 = math.atan((self.y-self.y0)/(self.x-self.x0))-math.atan(self.l2*math.sin(self.theta)/(self.l1+self.l2*math.cos(self.theta)))
        self.q2 = self.q1 + self.theta
        
        # Calculating the end points of the first arm from the point(x0,y0)
        self.x1 = self.x0 + self.l1*math.cos(self.q1)
        self.y1 = self.y0 + self.l1*math.sin(self.q1)
        
        # Calculating the end points of the first arm from the point(x1,y1)
        self.x2 = self.x1+self.l2*math.cos(self.q2)
        self.y2 = self.y1+self.l2*math.sin(self.q2)

        # Calculating the torques for both the arms
        self.t1 = -self.Fx*self.l1*math.sin(self.q1) + self.Fy*self.l1*math.cos(self.q1) + self.m1*g*self.l1*math.cos(self.q1)/2 + self.m2*g*self.l1*math.cos(self.q1)
        self.t2 = -self.Fx*self.l2*math.sin(self.q2) + self.Fy*self.l2*math.cos(self.q2) + self.m2*g*self.l2*math.cos(self.q2)/2
        
        self.its += 1

        # Adding the points in the coordinated in a set
        self.points.add((self.x1, self.y1, self.x2, self.y2, self.x, self.y))

    # draw function draws the image as per the user requirement on the basis of given coordinates
    def draw(self,image):
        for x1,y1,x2,y2,x,y in self.points:
            cv2.line(image,(self.x0,self.y0),(int(self.x1),int(self.y1)),(0,0,255),1)
            cv2.line(image,(int(self.x1),int(self.y1)),(int(self.x2),int(self.y2)),(0,255,0),1)
            cv2.circle(image,(int(self.x),int(self.y)),1,(255,0,0),-1)
        return image
```
