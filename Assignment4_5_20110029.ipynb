############## Question 9 ####################<br>
Code for finding desired angles given the end-effector position<br>
 Standford robot inverse kinematics

```python
import numpy as np
import math
xi,yi,zi = map(int,input("Enter the coordinates of the start point: ").split())
xf,yf,zf = map(int,input("Enter the coordinates of the end point: ").split())
```

 Length of robotic arm.

```python
l1 = 2 
l2 = 3
points = [] # Trajectories points
q_d = []  ## Vector of Desired angles 
for i in range(0,101):
    t = i/100
    x = xi+(xf-xi)*t
    y = yi+(yf-yi)*t
    z = zi+(zf-zi)*t
    
    q1d = math.atan(y/x)
    r = math.sqrt(x*2 + y*2)
    s = z - l1
    q2d = math.atan(s/r)
    q3d = math.sqrt(r*2+s*2)-l2
    
    q_d.append([q1d,q2d,q3d])
    points.append([x,y,z])
points = np.array(points)
q_d = np.array(q_d)
r = 0.01
theta_m_d = (1/r)*q_d
print(points)
print(q_d)
```
