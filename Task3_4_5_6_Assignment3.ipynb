# Task 3

```python
import numpy as np
import math

n = int(input(" Enter the number of links in your robot: "))
D_H = []
joints = [] 
for i in range(n):          
    a_i = int(input(f"Enter the value of a_i for link {i+1}: "))
    alpha_i = int(input(f"Enter the value of alpha_i for link {i+1}: "))
    d_i = int(input(f"Enter the value of d_i for link {i+1}: "))
    theta_i = int(input(f"Enter the value of theta_i for link {i+1}: "))
    D_H.append([a_i,alpha_i,d_i,theta_i])
    
k = input("Enter 'yes' if you want to provide the type of each joints and enter 'no' for all revolute joints as default: ")
if k=='yes':
    for i in range(n):  
        joint = input("Enter 'p' if joint is prismatic else enter 'r' if joint is revolute: ")
        joints.append(joint)
else:
     for i in range(n):  
        joint = 'r'
        joints.append(joint)
        
DH_par = np.array(D_H)
print(f'DH Parameters provided are:\n {DH_par} \n')

A = []

################ Homogenous transformation ###############
for i in range(n):
    r1 = [np.cos(DH_par[i,3]*math.pi/180),-1*np.sin(DH_par[i,3]*math.pi/180)*np.cos(DH_par[i,1]*math.pi/180),np.sin(DH_par[i,3]*math.pi/180)*np.sin(DH_par[i,1]*math.pi/180),DH_par[i,0]*np.cos(DH_par[i,3]*math.pi/180)]
    r2 = [np.sin(DH_par[i,3]*math.pi/180),np.cos(DH_par[i,3]*math.pi/180)*np.cos(DH_par[i,1]*math.pi/180),-1*np.cos(DH_par[i,3]*math.pi/180)*np.sin(DH_par[i,1]*math.pi/180),DH_par[i,0]*np.sin(DH_par[i,3]*math.pi/180)]
    r3 = [0,np.sin(DH_par[i,1]*math.pi/180),np.cos(DH_par[i,1]*math.pi/180),DH_par[i,2]]
    r4 = [0,0,0,1]
    A_i =np.array([r1,r2,r3,r4])
    A.append(A_i)
    
################ transformation for end- effector #############
T = []   # T is the list containing the value of A1,A1A2,A1A2A3,............., A1A2....An
Ti = np.identity(4)
for i in range(n):
    Ti = Ti@A[i]
    T.append(Ti)    
print(f'T{n}_0 is given by \n {T[n-1]} \n')

################ end- effector position #############
end_effector_pos = [T[n-1][0,3],T[n-1][1,3],T[n-1][2,3]]
print(f'\n end-effector position is \n {end_effector_pos} \n' )

O_n = np.array(end_effector_pos) # distance of end-effector from O_0
O_0 = np.array([0,0,0])
O = []                 # list contains the position of all origins
O.append(O_0)
j_v = []
j_w = []

for i in range(n):
    O_i = np.array([T[i][0,3],T[i][1,3],T[i][2,3]])
    O.append(O_i)
    
    Z_i = np.array([T[i][0,2],T[i][1,2],T[i][2,2]])   # Z_i is R_i_0 * k_i_0
    
    rv = np.cross(Z_i,(O_n - O[i]))             # rv denotes revolute joint value of element of velocity Jacobian
    pv = Z_i                                    # pv denotes prismatic joint value of element of velocity Jacobian
    if joints[i]=='p':
        j_v.append(pv)
        j_w.append(O_0)
    else:
        j_v.append(rv)
        j_w.append(pv)
        
Jv = np.transpose(np.array(j_v))    # velocity Jacobian
Jw = np.transpose(np.array(j_w))    # Angular velocity Jacobian
J = np.append(Jv,Jw,axis=0)         # Manipulator Jacobian
print(f"Manipulator Jacobian is \n {J}")



        
        
    
        
    
    
```
