Stanford Manipulator<br>
Considering only the links and not the wrist for calculation purpose.

```python
import numpy as np
import math
```

Length of all the link are l1,l2,l3 = 1

```python
l1 = 1
l2 = 1
l3 = 1
```

Coordinates of the end effector

```python
px = 1.13
py = 1.56
pz = 1.23
```

```python
r = math.sqrt(px**2 + py**2)
s = pz - l1
```

Calculation for theta1, theta2, and variable length d3

```python
theta1 = math.atan(py/px)
theta2 = math.atan(s/r)
```

Linear distance d3
