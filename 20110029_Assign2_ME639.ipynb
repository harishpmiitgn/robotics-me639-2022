```python
import numpy as np
import sympy as sym
import math
l1=int(input('Enter the length of arm 1: '))
l2=int(input('Enter the length of arm 2: '))
d3=int(input('Enter the distance d3: '))
# 
theta1, theta2 = sym.symbols('theta1 theta2 ')
x_pos=l1*sym.cos(theta1)+l2*sym.cos(theta2+theta1)
y_pos=l1*sym.sin(theta1)+l2*sym.sin(theta2+theta1)
z_pos=0
d_n_0=np.array([x_pos, y_pos, z_pos])
d_n_0=d_n_0.transpose()
t1=float(input('Enter the value of joint angle 1: '))*(math.pi)/180
t2=float(input('Enter the value of joint angle 2: '))*(math.pi)/180
```

Here I am using the approach is that ith column of Jv is differentiation of d^n_0 w.r.t theta_i

```python
x_pos_diff_1=sym.diff(x_pos,theta1)
x_pos_diff_1=x_pos_diff_1.subs([(theta1,t1),(theta2,t2)])
y_pos_diff_1=sym.diff(y_pos,theta1)
y_pos_diff_1=y_pos_diff_1.subs([(theta1,t1),(theta2,t2)])
x_pos_diff_2=sym.diff(x_pos,theta2)
x_pos_diff_2=x_pos_diff_2.subs([(theta1,t1),(theta2,t2)])
y_pos_diff_2=sym.diff(y_pos,theta2)
y_pos_diff_2=y_pos_diff_2.subs([(theta1,t1),(theta2,t2)])
```

```python
Jv=np.array([[x_pos_diff_1,x_pos_diff_2,0],[y_pos_diff_1,y_pos_diff_2,0],[0,0,-1]])
```

## Creating the Jw matrix

```python
Jw=sym.zeros(3,3)
for j in range(2):
    Jw[2,j]=1
Jw[2,2]=0
for i in range(2):
    for j in range(2):
        Jw[i,j]=0
#Combining both Jw and Jv
J=sym.zeros(6,3)
for i in range(6):
    for j in range(3):
        if i<=2:
            J[i,j]=Jw[i,j]
        else:
            J[i,j]=Jv[i-3,j]
```

 Finally printing the jacobian matrix

```python
print(J)
```

```python
import numpy as np
import sympy as sym
import math
l1=int(input('Enter the length of arm 1: '))
l2=int(input('Enter the length of arm 2: '))
l3=int(input('Enter the length of arm 3: '))
# 
theta1, theta2, theta3 = sym.symbols('theta1 theta2 theta3')
x_pos=l1*sym.cos(theta1)+l2*sym.cos(theta2+theta1)+l3*sym.cos(+theta1+theta2+theta3)
y_pos=l1*sym.sin(theta1)+l2*sym.sin(theta2+theta1)+l3*sym.sin(theta3+theta1+theta2)
z_pos=0
d_n_0=np.array([x_pos, y_pos, z_pos])
d_n_0=d_n_0.transpose()
t1=float(input('Enter the value of joint angle 1: '))*(math.pi)/180
t2=float(input('Enter the value of joint angle 2: '))*(math.pi)/180
t3=float(input('Enter the value of joint angle 3: '))*(math.pi)/180
```

## Here I am using the approach is that ith column of Jv is differentiation of d^n_0 w.r.t theta_i.#####

```python
x_pos_diff_1=sym.diff(x_pos,theta1)
x_pos_diff_1=x_pos_diff_1.subs([(theta1,t1),(theta2,t2),(theta3,t3)])
y_pos_diff_1=sym.diff(y_pos,theta1)
y_pos_diff_1=y_pos_diff_1.subs([(theta1,t1),(theta2,t2),(theta3,t3)])
x_pos_diff_2=sym.diff(x_pos,theta2)
x_pos_diff_2=x_pos_diff_2.subs([(theta1,t1),(theta2,t2),(theta3,t3)])
y_pos_diff_2=sym.diff(y_pos,theta2)
y_pos_diff_2=y_pos_diff_2.subs([(theta1,t1),(theta2,t2),(theta3,t3)])
x_pos_diff_3=sym.diff(x_pos,theta3)
x_pos_diff_3=x_pos_diff_3.subs([(theta1,t1),(theta2,t2),(theta3,t3)])
y_pos_diff_3=sym.diff(y_pos,theta3)
y_pos_diff_3=y_pos_diff_3.subs([(theta1,t1),(theta2,t2),(theta3,t3)])
```

```python
Jv=np.array([[x_pos_diff_1,x_pos_diff_2,x_pos_diff_3],[y_pos_diff_1,y_pos_diff_2,y_pos_diff_3],[0,0,0]])
```

## Creating the Jw matrix

```python
Jw=sym.zeros(3,3)
for j in range(3):
    Jw[2,j]=1
for i in range(2):
    for j in range(3):
        Jw[i,j]=0
#### Combining Jw and Jv to get J.
J=sym.zeros(6,3)
for i in range(6):
    for j in range(3):
        if i<=2:
            J[i,j]=Jw[i,j]
        else:
            J[i,j]=Jv[i-3,j]
```

```python
 # Printing the Jacobian matrix           
print(J)
```
