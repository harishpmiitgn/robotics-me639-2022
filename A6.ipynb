```python
import numpy as np
from math import cos,sin,acos,atan,pi
```

```python
print("Enter the lengths of 1st and 2nd arm - ")
a1 = int(input())
a2 = int(input())
```

```python
print("Enter the distance of the tip of the tool from the wrist:")
d4 = int(input())
```

```python
r11 = 1.6
r12 = 1.2
```

```python
d = np.array([[0.95], [2.3], [4]])
```

Calculation of joint angles q1, q2, q3 and q4

```python
q2 = acos((d[0]*2 + d[1]**2 - a1**2 - a2**2)/(2*a1*a2))
q1 = atan(d[1]/d[0])- atan((a2*sin(q2))/(a1 + a2*cos(q2)))
q3 = - (d[2] + d4)
q4 = q1 + q2 - atan(r11/r12)
```

```python
print('The calculated joint variables are (in degrees) - ')
print('Q1 - ' + str(q1*180/pi))
print('Q2 - ' + str(q2*180/pi))
print('Q3 - ' + str(q3*180/pi))
print('Q4 - ' + str(q4*180/pi))
```

End effector coordinates are given as following

```python
d0_4 = np.array([[a1*cos(q1) + a2*cos(q1+q2)], [a1*sin(q1) + a2*sin(q1+q2)], [-q3-d4]])
```

```python
print('The initial position of the end effector was')
print(d0_4[0][0])
print(d0_4[1][0])
print(d0_4[2][0])
```
